## Primary analysis of the data show the following: 
    * Missing values in Height and Weight that can be filled.
    * Medals has over 85% of the data missing which is a problem
    * Age has 9474 missing values that can be imputed.
    * Notes Will be dropped.
    * Replace medals with no entries to No medal

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Height'].fillna(df['Height'].mean(),inplace=True)
df['Weight'].fillna(df['Weight'].mean(),inplace=True)
df['Medal'].fillna('No Medal',inplace=True)

```

# Now the Data is ready for the Database


In the database. A star schema was used because of its simplicity and represntative abilites for the data.


### Table 1 : Games

```python
table_games = df[['Games', 'Year', 'Season', 'City']].drop_duplicates()
table_games['Games_ID'] = table_games.index + 1
```

```python
table_games.head()
```

### Table 2 : Events

```python
table_events = df[["Event", "Sport"]].drop_duplicates()
table_events['Event_ID'] = table_events.index + 1
```

```python
table_events.head()
```

### Table 3 : Athletes

```python
table_athletes = df[['ID', 'Name', 'Sex', 'Height', 'Weight', 'Team', 'NOC']].drop_duplicates()
table_athletes = table_athletes.rename(columns={'ID': 'AthleteID'})
```

```python
table_athletes.head()
```

### Table 4 : Countries

```python
table_countries = df[['NOC', 'region']].drop_duplicates()
table_countries['Country_ID'] = table_countries.index + 1
```

```python
table_countries.head()
```

### Table 5 : Results

```python
# Create the table_results directly from df without merging
table_results = df[['ID', 'Event', 'Sport', 'Games', 'Year', 'Season', 'City', 'Age', 'Medal']].drop_duplicates()

# Map Event and Sport to Event_ID
event_id_map = table_events.set_index(['Event', 'Sport'])['Event_ID'].to_dict()
table_results['Event_ID'] = table_results.apply(lambda row: event_id_map.get((row['Event'], row['Sport'])), axis=1)

# Map Games data to Games_ID
games_id_map = table_games.set_index(['Games', 'Year', 'Season', 'City'])['Games_ID'].to_dict()
table_results['Games_ID'] = table_results.apply(lambda row: games_id_map.get((row['Games'], row['Year'], row['Season'], row['City'])), axis=1)

# Rename ID to Athlete_ID to match with table_athletes
table_results = table_results.rename(columns={'ID': 'Athlete_ID'})

# Drop the columns we no longer need
table_results = table_results[['Athlete_ID', 'Event_ID', 'Games_ID', 'Age', 'Medal']]

# Add Result_ID
table_results['Result_ID'] = table_results.index + 1

# Reorder columns to match the schema
table_results = table_results[['Result_ID', 'Athlete_ID', 'Event_ID', 'Games_ID', 'Age', 'Medal']]
```

```python
table_results.head()
```

# Create SQLite Database & Tables

In this section, we'll create a SQLite database based on our data model. The schema includes:
- COUNTRIES table
- ATHLETES table
- EVENTS table
- GAMES table
- RESULTS table

```python
# Create a new database connection
conn = sqlite3.connect('olympics.db')
cursor = conn.cursor()

print(" Connected to a SQLite database.")
```

### Creating Tables 

```python
# Create COUNTRIES table
cursor.execute('''
CREATE TABLE COUNTRIES (
    Country_ID INTEGER PRIMARY KEY,
    NOC TEXT UNIQUE,
    Region TEXT,
    Notes TEXT
)
''')

# Create ATHLETES table
cursor.execute('''
CREATE TABLE ATHLETES (
    ID INTEGER PRIMARY KEY,
    Name TEXT,
    Sex TEXT,
    Height INTEGER,
    Weight REAL,
    Team TEXT,
    NOC TEXT,
    Country_ID INTEGER,
    FOREIGN KEY (Country_ID) REFERENCES COUNTRIES(Country_ID)
)
''')

# Create EVENTS table
cursor.execute('''
CREATE TABLE EVENTS (
    Event_ID INTEGER PRIMARY KEY,
    Event TEXT UNIQUE,
    Sport TEXT
)
''')

# Create GAMES table
cursor.execute('''
CREATE TABLE GAMES (
    Games_ID TEXT PRIMARY KEY,
    Games TEXT,
    Year INTEGER,
    Season TEXT,
    City TEXT
)
''')

# Create RESULTS table
cursor.execute('''
CREATE TABLE RESULTS (
    Result_ID INTEGER PRIMARY KEY,
    Athlete_ID INTEGER,
    Event_ID INTEGER,
    Games_ID TEXT,
    Age INTEGER,
    Medal TEXT,
    FOREIGN KEY (Athlete_ID) REFERENCES ATHLETES(ID),
    FOREIGN KEY (Event_ID) REFERENCES EVENTS(Event_ID),
    FOREIGN KEY (Games_ID) REFERENCES GAMES(Games_ID)
)
''')

conn.commit()
```

Since we sucessfuly created the tables. we'll prepare the data for each table from our dataframe.

```python
# Insert data into COUNTRIES table
table_countries.to_sql('COUNTRIES', conn, if_exists='replace', index=False)

# Insert data into EVENTS table
table_events.to_sql('EVENTS', conn, if_exists='replace', index=False)

# Insert data into GAMES table
table_games.to_sql('GAMES', conn, if_exists='replace', index=False)

# Insert data into ATHLETES table
table_athletes.to_sql('ATHLETES', conn, if_exists='replace', index=False)

# Insert data into RESULTS table
table_results.to_sql('RESULTS', conn, if_exists='replace', index=False)

conn.commit()
```
